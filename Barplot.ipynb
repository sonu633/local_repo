```python
import seaborn as sns
import matplotlib as plt
import pandas as pd

df=sns.load_dataset("titanic")
df.head()
```

```python
sns.barplot(x="who",y="fare",data=df)
plt.show()
```

```python
sns.barplot(x="who",y="fare",data=df,color="magenta")
plt.show()
```

```python
sns.barplot(x="sex",y="fare",data=df,color="magenta",hue="who",palette="spring",saturation=0.2)
plt.show()
```

```python
from numpy import mean
sns.barplot(x="fare",y="who",data=df,ci=False,color="magenta",saturation=0.2,estimator=mean)
plt.show()
```

```python
from numpy import median
sns.barplot(x="fare",y="who",data=df,estimator=median)
plt.show()
```

```python
sns.barplot(x="fare",y="who",data=df,facecolor=[1,1,1,0],edgecolor="0.2")
plt.show()
```

```python
sns.barplot(x="fare",y="who",data=df ,order=["child","woman","man"])
plt.show()
```
