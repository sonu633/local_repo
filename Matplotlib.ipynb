## Matplotlib
## pieplot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cars=["BMW","Ferrari","Zip","scorpio"]
data=[23,34,24,25]
plt.figure(figsize=(10,5))
colors = ['yellow', 'green', 'blue', 'orange']
explode = [0.3, 0, 0, 0]
plt.pie(data,labels=cars, autopct='%d%%',colors=colors,explode =explode,wedgeprops={'edgecolor': 'black', 'linewidth': 2})
plt.show()
```
