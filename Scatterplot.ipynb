```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df =sns.load_dataset("titanic")
df.head()
```

```python
sns.scatterplot(x="who",y="sex",data=df)
plt.show
```

```python
sns.scatterplot(x="pclass",y="age",data=df)
plt.show()
```

```python
sns.scatterplot(x="pclass",y="age",data=df,hue="survived",palette="GnBu")
plt.show()
```

```python
sns.scatterplot(x="who",y="sex",data=df,size="class")
plt.show()
```

```python
sns.scatterplot(x="pclass",y="age",data=df,alpha=0.5)
plt.show()
```

```python
sns.scatterplot(x="who",y="sex",data=df,marker="v")
plt.show
```
