```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd

df= sns.load_dataset("iris")
df.head()
```

```python
sns.pairplot(data=df,height=1.5)
plt.show()
```

```python
sns.pairplot(data=df,height=1.5,hue=("species"),palette ="viridis")
plt.show()
```

```python
sns.pairplot(data=df,height=1.5,hue=("species"),palette ="viridis",diag_kind="hist")
plt.show()
```

```python
sns.pairplot(data=df,height=1.5,hue=("species"),palette ="viridis",diag_kind="kde")
plt.show()
```

```python
sns.pairplot(data=df,height=1.5,plot_kws ={"edgecolor":"yellow"})
plt.show()
```
